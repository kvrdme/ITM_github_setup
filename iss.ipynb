```python
import matplotlib.pyplot as plt
import requests
```

```python
url = 'http://api.open-notify.org/iss-now.json'
```

```python
import requests
import time

data_points = []
num_data_points = 100
interval = 10  #seconds

#making API requests
for i in range(num_data_points):
    response = requests.get(url)
    data = response.json()
    timestamp = data['timestamp']
    latitude = data['iss_position']['latitude']
    longitude = data['iss_position']['longitude']
    data_points.append({'timestamp': timestamp, 'latitude': latitude, 'longitude': longitude})
    time.sleep(interval)

#printing the generated data points
for data_point in data_points:
    print(data_point)
```

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature

#extracting latitude and longitude coordinates
latitudes = [point['latitude'] for point in data_points]
longitudes = [point['longitude'] for point in data_points]

#creating the map
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

#map to cover entire globe
ax.set_global()

#coastlines, countries, and borders
ax.coastlines()
ax.add_feature(cfeature.BORDERS)

#plotting the path of the ISS on the map
ax.plot(longitudes, latitudes, marker='o', markersize=1, color='blue', transform=ccrs.PlateCarree())

#plot title
ax.set_title('Path of the International Space Station: 07/01/23, 9PM')

#showing map and plot
plt.show()
```
